```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Функція для обчислення інтерполяційної функції Лагранжа
def lagrange_interpolation(x, y, x_test):
    n = len(x)
    p = np.zeros(n)
    for i in range(n):
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_test - x[j]) / (x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p)

# Задана функція
x = np.array([-2., 0., 2., 4.], dtype=float)
y = np.array([-5., 1., -9., 13.], dtype=float)

# Точки, для яких проводиться інтерполяція
x_test = [-2, -1, 0, 3, -1.5, -0.5, 1, 2]

# Виконання інтерполяції та вивід результатів
for point in x_test:
    interpolated_value = lagrange_interpolation(x, y, point)
    print(f"Значення функції у точці x={point}: {interpolated_value}")

# Точки, за якими будуємо графік
x_new = np.linspace(np.min(x), np.max(x), 100)
y_new = [lagrange_interpolation(x, y, i) for i in x_new]

# Графік інтерполяційної функції Лагранжа
plt.plot(x, y, 'mo', x_new, y_new, 'g-')
plt.scatter(x_test, [lagrange_interpolation(x, y, i) for i in x_test], color='red')  # Додавання червоних точок для нових точок
plt.title('Інтерполяційний багаточлен Лагранжа')
plt.grid(True)
plt.show()

# Перевірка
f = lagrange(x, y)
fig = plt.figure(figsize=(7, 5))
plt.plot(x, y, 'bo', x_new, f(x_new), 'y-')
plt.scatter(x_test, [lagrange_interpolation(x, y, i) for i in x_test], color='red')  # Додавання червоних точок для нових точок
plt.title('Інтерполяційний багаточлен Лагранжа (перевірка)')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
